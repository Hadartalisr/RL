```python
import math

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.special import comb
```

```python

# Set the parameters for the binomial distribution
n = 20
p = 0.5

# Define the range of k values
k_values = np.arange(0, n+1)  # extended range for visualization purposes
normalized_k_values = k_values / n

# Define f_1 as the binomial PMF
def f_1(k, n):
    return comb(n, k, exact=True) / math.pow(2, n)


f_1_values = [f_1(k, n) for k in k_values]


def f_3(k, n): 
    if k < n/2:
        return 0
    return 2 * f_1(k, n)


f_3_values = [f_3(k, n) for k in k_values]

# Create the plot
plt.figure(figsize=(10, 6))
plt.stem(normalized_k_values, f_1_values, basefmt=" ",  label='f1')
plt.stem(normalized_k_values - 0.5, f_1_values, basefmt=" ", linefmt='g-', markerfmt='go',  label='f2')
plt.stem(normalized_k_values - 0.5, f_3_values, basefmt=" ", linefmt='r-', markerfmt='ro', label='f3')
# Add labels and title
plt.xlabel('k/n')
plt.ylabel('Probability')
plt.title('Visualization of Binomial PMF Functions')
plt.legend()
plt.show()


```

```python

# Parameters for the normal distribution
mu = 0  # mean
sigma = n/4  # standard deviation

# Generate points for the normal distribution
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
normal_distribution = (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Generate points for the folded normal distribution
folded_normal_distribution = (1 / (np.sqrt(2 * np.pi) * sigma)) * (np.exp(-0.5 * ((x - mu) / sigma)**2) + np.exp(-0.5 * ((x + mu) / sigma)**2))
# Zero out values where x < 0
folded_normal_distribution = np.where(x < 0, 0, folded_normal_distribution)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x/n, normal_distribution, label='Normal Distribution', color='blue')
plt.plot(x/n, folded_normal_distribution, label='Folded Normal Distribution', color='red')
plt.title('Normal Distribution vs Folded Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
